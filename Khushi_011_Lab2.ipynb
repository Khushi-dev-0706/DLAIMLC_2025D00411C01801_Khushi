```python
import tensorflow as tf
import numpy as np
```

```python
# Step 1: Given Data
X = [75, 65, 80]          # Inputs
w = [4, 3, 5]             # Initial weights
b = -1000                 # Bias
eta = 0.01                # Learning rate
y = 1                     # Target output (Pass = 1)

print("BEFORE LEARNING")
```

```python
# Step 2: Net input
z = (w[0]*X[0]) + (w[1]*X[1]) + (w[2]*X[2]) + b
print("Net input z =", z)
```

```python
def step_function(z):
    if z >= 0:
        return 1
    else:
        return 0
```

```python
# Step 3: Activation
y_pred = step_function(z)
print("Prediction =", "Pass" if y_pred == 1 else "Fail")
```

```python
# Step 4: Error
error = y - y_pred
print("Error =", error)

# Step 5: Weight update
if error != 0:
    w[0] = w[0] + eta * error * X[0]
    w[1] = w[1] + eta * error * X[1]
    w[2] = w[2] + eta * error * X[2]

# Step 6: Bias update
    b = b + eta * error
```

```python
print("\nAFTER LEARNING")

z_new = (w[0]*X[0]) + (w[1]*X[1]) + (w[2]*X[2]) + b
print("Net input z =", z_new)

y_pred_new = step_function(z_new)
print("Prediction =", "Pass" if y_pred_new == 1 else "Fail")

print("\nUpdated Weights:", w)
print("Updated Bias:", b)
```
